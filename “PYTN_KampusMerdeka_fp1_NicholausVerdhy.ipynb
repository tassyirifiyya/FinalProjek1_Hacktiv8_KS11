```python
new_cat = ['source',
 'destination',
 'cab_type',
 'name',
 'short_summary',
 'icon','price']

df_cat = df[new_cat]
df_cat.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

df_cat_encode= df_cat.copy()
for col in df_cat_encode.select_dtypes(include='O').columns:
    df_cat_encode[col]=le.fit_transform(df_cat_encode[col])
```

```python
df_cor = df.loc[:,['cab_type','name','price','distance','surge_multiplier']]

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 17))
ax = sns.heatmap(df_cor.corr(),annot=True, cmap='viridis', linewidths=.5, linecolor='black', square=True)
```

```python
df_banding_harga = df.loc[:,["cab_type","price"]]
df_banding_harga
```

```python
grouped = df_banding_harga.groupby("cab_type")
mean_dfbanding_harga = grouped["price"].mean()
```

```python
mean_dfbanding_harga
```

```python
plt.bar(mean_dfbanding_harga.index,mean_dfbanding_harga.values)
plt.xlabel('cab_type')
plt.ylabel('Rata-rata Harga')
plt.title('Perbandingan Rata-rata Harga berdasarkan cab_type')
plt.show()
```
